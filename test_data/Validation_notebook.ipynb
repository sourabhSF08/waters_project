```python
user17985= data.query('accountsapid==17985')
user17985
```

- As we can see users have transactions in 2019 ,2020,2021 


```python
user17985 = user17985.sort_values('OrderCreateDate')
DiffDate = user17985['OrderCreateDate'] -user17985['OrderCreateDate'].shift(1)

user17985['DiffDate'] = DiffDate
print(user17985[['OrderCreateDate','DiffDate']])

print("Difference between dates\n",DiffDate)

print("\nMedian",median(DiffDate[1:]).days)
print('Median with 3 years cycle ',median(DiffDate[1:]).days*3)

print('\nLast Order Date\n', user17985['OrderCreateDate'][-1:])

print("\nNow Acording to rule given in ChurnTestCaseSimulation\nwe will see whether 3year cycle date has come before or after \nlast purchase date")
print("\nIf last purchase date has come before 3 year cycle date then we will churn out that user \nelse churn in ")

if median(DiffDate[1:]).days*3 >= (date.today()-date(user17985['OrderCreateDate'][-1:].dt.year,user17985['OrderCreateDate'][-1:].dt.month,user17985['OrderCreateDate'][-1:].dt.day )).days:
    print('\n\nWillnot_churn')
else:
    print('\n\nWill_churn')
```

- So after Cross checking We can see all our result is perfect
